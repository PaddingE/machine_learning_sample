```python
import numpy as np
import matplotlib.pyplot as plt

## 데이터 설정
np.random.seed(42)
arr_input = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.5)
arr_target = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.25)
l_rate = 0.001
a = 0.0
b = 0.0
delta_a = 0.0
delta_b = 0.0

# g_x = arr_target - ((a * arr_input) + b)
# f_x = np.sum(g_x ** 2) / len(arr_input)
```

```python
## 오차가 적은 선형 방정식을 찾기 위한 부분
for _ in range(len(arr_input)):
    ## 선형 방정식의 계수와 절편을 구하기 위한 델타값
    delta_a = -2 * np.sum(arr_input * (arr_target - ((a * arr_input) + b))) / len(arr_input)
    delta_b = -2 * np.sum(arr_target - ((a * arr_input) + b)) / len(arr_input)
    
    ## 계수와 절편 구하기
    a = a - l_rate * delta_a
    b = b - l_rate * delta_b

    g_x = arr_target - ((a * arr_input) + b)
    f_x = np.sum(g_x ** 2) / len(arr_input)
```
