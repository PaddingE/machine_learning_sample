```python
import numpy as np
import matplotlib.pyplot as plt
## 만든 class import
from Linear import Linear

## 데이터 설정
np.random.seed(42)
arr_input = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.5)
arr_target = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.25)
```

```python
## 객체 생성
c_ln = Linear()

## 훈련함수 사용
c_ln.fit(arr_input, arr_target)
```

```python
## 선형 방정식 그래프로 확인
plt.scatter(arr_input,arr_target)
plt.plot(arr_input,c_ln.predict(arr_input))
plt.xlabel('input')
plt.ylabel('target')
plt.show()
```

```python
## 추가학습 함수 사용
c_ln.partial_fit(arr_input, arr_target)
```

```python
## 추가 학습 후 선형 방정식 그래프화
plt.scatter(arr_input,arr_target)
plt.plot(arr_input,c_ln.predict(arr_input))
plt.xlabel('input')
plt.ylabel('target')
plt.show()
```
